```python
from tkinter import *

window = Tk()
window.configure(bg="#b16ae8")
window.geometry("500x600")
window.title("Submission Form")

lbl = Label(
    window,
    text="Submission Form",
    width=15,
    height=4,
    font=(
        "Arial",
        10,
    ),
    bg="#b16ae8",
    fg="white",
    bd=2,
)


def submit():
    username = name.get()
    userage = age.get()
    usermobile = mobile.get()
    if userage.isdigit() and usermobile.isdigit():
        if len(usermobile) == 10:
            if int(userage) >= 18:
                print(f'Hey {username} You are eligible for voting')
                name.config(state=DISABLED)
                age.config(state=DISABLED)
                mobile.config(state=DISABLED)
            else:
                print(f'Hey {username} You are not eligible for voting')
                name.config(state=DISABLED)
                age.config(state=DISABLED)
                mobile.config(state=DISABLED)
        else:
            print('Mobile number should be equal to 10 digits excluding country code')
    else:
        print('Please enter only numerical values in age and mobile')

def dlt():
    name.delete(0,END)
    age.delete(0,END)
    mobile.delete(0,END)


def bckspc():
    name.delete(len(name.get()) - 1,END)
    age.delete(len(age.get()) - 1,END)
    mobile.delete(len(mobile.get()) - 1,END)

lbl_name = Label(window,text='Name')
name = Entry(window, name="name", width=30)
age = Entry(window, name="age", width=30)
mobile = Entry(window, name="mobile", show="*", width=30)

lbl.pack(padx=20, pady=20)
lbl_name.pack()
name.pack(padx=10, pady=10)
age.pack(padx=10, pady=10)
mobile.pack(padx=10, pady=10)

submit = Button(window, text="Submit",command=submit,width=30)
delete = Button(window, text="Reset", command=dlt, width=30)
bck = Button(window, text="Backspace", command=bckspc, width=30)

submit.pack(padx=10,pady=10)
delete.pack(padx=10,pady=10)
bck.pack(padx=10,pady=10)
window.mainloop()
```
