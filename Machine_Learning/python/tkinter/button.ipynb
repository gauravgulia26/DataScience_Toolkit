```python
from tkinter import *

window = Tk()
window.title("Buttons")
window.geometry("500x500")
window.configure(bg="#0381ff")


# adding a image to the button

pht = PhotoImage(file='img_2.png')
# how to check which button is pressed
def which_button(button_press):
    print(button_press)


btn = Button(
    window,
    text="Button 1",
    bd=4,
    font=("roboto", 14, "bold"),
    bg="black",
    fg="lime",
    height=4,
    width=20,
    relief="raised",
    activebackground="#3d3432",
    activeforeground="red",
    cursor='hand2',
    command=lambda: which_button('button 1'),
    image=pht
)

btn.pack(padx=10, pady=10)

btn = Button(
    window,
    text="Button 2",
    bd=4,
    font=("roboto", 14, "bold"),
    bg="black",
    fg="lime",
    height=4,
    width=20,
    relief="raised",
    activebackground="#3d3432",
    activeforeground="red",
    cursor='hand2',
    command=lambda: which_button('button 2')
)

btn.pack(padx=10, pady=10)


# text_update function
def text_updation(language):
    text.delete(0, END)
    text.insert(0, language)


# Entry Box
text = Entry(window, width=30, bg="White")
text.pack(pady=10)

# create buttons
button_dict = {}

words = ["Python", "Java", "R", "JavaScript"]
for lang in words:

    # pass each button's text to a function
    def action(x=lang):
        return text_updation(x)

    # create the buttons
    button_dict[lang] = Button(window, text=lang, command=action)
    button_dict[lang].pack(pady=10)

window.mainloop()
```

```python
from tkinter import Tk, Label
from PIL import Image, ImageTk

# Create the main window
window = Tk()
window.configure(bg="black")
window.geometry("400x400")
window.title("Hey")

# Open and resize the image
image = Image.open("img_2.png")
image = image.resize((50, 50))

# Convert the resized image to a PhotoImage object
photo = ImageTk.PhotoImage(image)

# Create a label with the image
label = Label(
    window, text="Hey", image=photo, bg="white", height=400, width=400, compound="top"
)
label.image = photo  # Keep a reference to prevent garbage collection
label.pack()

# Start the Tkinter event loop
window.mainloop()
```
