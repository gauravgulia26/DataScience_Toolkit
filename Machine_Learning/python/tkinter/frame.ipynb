```python
from tkinter import *

# Create the main window
window = Tk()
window.geometry("500x600")
window.title("Framing in tkinter")

# Create a frame
frm = Frame(window)  # Attach the frame to the main window
frm.pack()  # Pack the frame into the window

# Create a label inside the frame
lbl = Label(frm, text="Hey, I am a label")  # Attach the label to the frame
lbl.pack()  # Pack the label into the frame

# Run the main event loop
window.mainloop()
```
