# Axis in Numpy
axis 0 (Vertical Axis / Row-wise):

    Refers to the first dimension.
    Think of it as operating "down" the rows.
    For example, in a 2D array, axis 0 processes the elements column by column.

axis 1 (Horizontal Axis / Column-wise):

    Refers to the second dimension.
    Think of it as operating "across" the columns.
    For example, in a 2D array, axis 1 processes the elements row by row.

```python
np.arange(100).reshape(20, 5).sum(axis=0)
```

```python
np.arange(100).reshape(20, 5).sum(axis=1)
```

```python
np.arange(100).reshape(20, 5).mean(axis=0)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
class Trignometry:
    def __init__(self, data: np.ndarray):
        self.tensor = data

    def sin_square(self):
        return np.square(np.sin(self.tensor)) + np.square(np.cos(self.tensor))
```

```python
theta = Trignometry(data)
```

```python
theta.sin_square()
```
