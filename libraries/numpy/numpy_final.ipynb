```python
import numpy as np
import matplotlib.pyplot as plt

# Define the tensors
x1 = np.random.rand(10)
cx2 = np.random.rand(10)
y1 = np.random.rand(10)
y2 = np.random.rand(10)

# Calculate the Euclidean distances
distances = np.sqrt((x1 - cx2) ** 2 + (y1 - y2) ** 2)

# Find maximum and minimum distances
max_idx = np.argmax(distances)
min_idx = np.argmin(distances)

# Plot the distances
plt.figure(figsize=(8, 4))
plt.plot(distances, marker="o", linestyle="-", color="b", label="Euclidean Distance")

# Annotate each point
for i, dist in enumerate(distances):
    plt.annotate(
        f"{dist:.2f}",
        (i, dist),
        textcoords="offset points",
        xytext=(0, 10),
        ha="center",
        fontsize=10,
    )

# Highlight maximum and minimum distances
plt.scatter(max_idx, distances[max_idx], color="red", label="Max Distance", zorder=5)
plt.scatter(min_idx, distances[min_idx], color="green", label="Min Distance", zorder=5)

# Add annotations for max and min
plt.annotate(
    f"Max: {distances[max_idx]:.2f}",
    (max_idx, distances[max_idx]),
    textcoords="offset points",
    xytext=(-20, 15),
    fontsize=12,
    color="red",
    arrowprops=dict(arrowstyle="->", color="red"),
)
plt.annotate(
    f"Min: {distances[min_idx]:.2f}",
    (min_idx, distances[min_idx]),
    textcoords="offset points",
    xytext=(-20, -25),
    fontsize=12,
    color="green",
    arrowprops=dict(arrowstyle="->", color="green"),
)

# Add titles and labels
plt.title("Euclidean Distances Between Points", fontsize=16)
plt.xlabel("Index", fontsize=14)
plt.ylabel("Distance", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)
plt.show()
```
