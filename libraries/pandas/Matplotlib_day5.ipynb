```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv(
    "/home/skynet/Documents/DataScience_Toolkit/libraries/pandas/data/sadak_yojna.csv"
)
```

```python
df.groupby("PMGSY_SCHEME")["NO_OF_ROAD_WORK_SANCTIONED"].mean().sort_values(
    ascending=False
).plot(kind="bar", color="green")
plt.show()
```

```python
df.groupby("PMGSY_SCHEME")["NO_OF_ROAD_WORK_SANCTIONED"].mean().sort_values(
    ascending=False
).plot(kind="barh", color="green")
plt.show()
```

```python
# plt.style.available
title_font = {
    "family": "sans-serif",
    "style": "italic",
    "weight": "bold",
    "color": "darkblue",
    "size": 18,
}
```

```python
plt.figure(figsize=(8, 6))
explode = [0.05] * len(
    df.groupby("PMGSY_SCHEME")["NO_OF_ROAD_WORK_SANCTIONED"]
    .mean()
    .sort_values(ascending=False)
)
colors = plt.cm.Paired(
    range(
        len(
            df.groupby("PMGSY_SCHEME")["NO_OF_ROAD_WORK_SANCTIONED"]
            .mean()
            .sort_values(ascending=False)
        )
    )
)
df.groupby("PMGSY_SCHEME")["NO_OF_ROAD_WORK_SANCTIONED"].mean().sort_values(
    ascending=False
).plot(
    kind="pie",
    color="green",
    autopct="%1.1f%%",
    startangle=90,
    explode=explode,
    textprops={"fontsize": 12},
)
plt.ylabel("")
plt.legend(loc="lower right")

plt.tight_layout()
plt.title(
    "Road Sanctioned Distribution Across Various Schemes",
    loc="center",
    fontsize=10,
    pad=30,
    fontdict=title_font,
)
plt.legend(
    title="Schemes",
    loc="center left",
    bbox_to_anchor=(1, 0.5),  # Place legend outside the plot
    fontsize=10,
)
plt.style.use("dark_background")
plt.show()
```

```python
q3 = df["LENGTH_OF_ROAD_WORK_SANCTIONED_KM"].quantile(0.75)
q1 = df["LENGTH_OF_ROAD_WORK_SANCTIONED_KM"].quantile(0.25)
```

```python
iqr = q3 - q1
upper = q3 + 1.5 * iqr
```

```python
pd.Series(df["LENGTH_OF_ROAD_WORK_SANCTIONED_KM"].values.clip(0, upper)).plot(
    kind="kde"
)
```

```python
pd.Series(df["LENGTH_OF_ROAD_WORK_SANCTIONED_KM"].values.clip(0, upper)).plot(
    kind="box"
)
# plt.axhline(upper)
plt.annotate(text="Q3", xy=(0.5, 400))
plt.annotate(text="Q1", xy=(0.5, 110))
plt.annotate(text="Upper", xy=(0.5, upper))
plt.show()
```

```python
x = np.arange(1, 1000)
y = np.linspace(1, 10, 999)
```

```python
synth_data = pd.Series(data=x, index=y)
mu = synth_data.mean()
mu1 = mu + synth_data.std()
mu2 = mu + 2 * synth_data.std()
mu3 = mu + 3 * synth_data.std()
synth_data.plot(kind="kde")
plt.axvline(mu, label="Mean")
plt.axvline(mu1, label="First Standard Devaition", color="cyan")
plt.axvline(mu2, label="Second STD", color="red")
plt.axvline(mu3, label="Third STD", color="blue")
plt.legend()
plt.show()
```
