```python
import warnings
import matplotlib.pyplot as plt

# plt.polar([0, 1], [0, 2])

warnings.filterwarnings("ignore")
df[
    [
        "NO_OF_ROAD_WORKS_COMPLETED",
        "EXPENDITURE_OCCURED_LAKHS",
        "LENGTH_OF_ROAD_WORK_BALANCE_KM",
    ]
].plot(
    kind="scatter",
    x="NO_OF_ROAD_WORKS_COMPLETED",
    y="EXPENDITURE_OCCURED_LAKHS",
    s="LENGTH_OF_ROAD_WORK_BALANCE_KM",
    c="LENGTH_OF_ROAD_WORK_BALANCE_KM",
    colormap="viridis",
)
plt.title("My Plot")
plt.annotate(
    text="This is the Maximum",
    xy=(df["NO_OF_ROAD_WORKS_COMPLETED"].max(), df["EXPENDITURE_OCCURED_LAKHS"].max()),
    textcoords="figure points",
    xytext=(100, 300),
    arrowprops=dict(facecolor="green", arrowstyle="->", linewidth=1.5),
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="white", facecolor="gray"),
)
plt.annotate(
    text="This is the Average",
    xy=(
        df["NO_OF_ROAD_WORKS_COMPLETED"].mean(),
        df["EXPENDITURE_OCCURED_LAKHS"].mean(),
    ),
    textcoords="figure points",
    xytext=(250, 100),
    arrowprops=dict(facecolor="green", arrowstyle="->", linewidth=1.5),
    bbox=dict(boxstyle="round,pad=0.3", edgecolor="white", facecolor="gray"),
)
plt.tight_layout()
plt.show()
```
