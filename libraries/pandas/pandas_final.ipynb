```python
import numpy as np
import pandas as pd
from scipy.stats import norm

# Define probabilities (percentiles)
probabilities = np.linspace(0.01, 0.99, 100)

# Compute theoretical quantiles for a standard normal distribution
quantiles = norm.ppf(probabilities)

# Create a DataFrame for better representation
quantiles_table = pd.DataFrame(
    {"Probability (p)": probabilities, "Theoretical Quantile (Z)": quantiles}
)

quantiles_table
```

```python
# %pip install statsmodels
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.gofplots import qqplot

# Generate synthetic data from a normal distribution
np.random.seed(42)  # For reproducibility
synthetic_data = np.random.normal(
    loc=0, scale=1, size=1000
)  # Mean=0, StdDev=1, 1000 samples

# Create QQ plot
fig = plt.figure(figsize=(8, 8))
qqplot(synthetic_data, line="s", ax=plt.gca())
plt.title("QQ Plot of Synthetic Normal Data", fontsize=14)
plt.show()
```
