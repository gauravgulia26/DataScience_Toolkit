```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "/home/skynet/Documents/DataScience_Toolkit/libraries/pandas/data/sadak_yojna.csv"
)
```

```python
df.head()
```

```python
df[["STATE_NAME", "PMGSY_SCHEME"]]
```

```python
# Find hte scheme with the most value across all the state
pd.crosstab(df["PMGSY_SCHEME"], df["STATE_NAME"]).mean(axis=1).sort_values(
    ascending=False
).plot(kind="bar")
plt.show()
```

```python
plt.style.available
```

```python
# plt.figure(figsize=(20, 20))
(pd.crosstab(df["STATE_NAME"], df["PMGSY_SCHEME"], normalize="index") * 100).plot(
    kind="bar", stacked=True, figsize=(14, 5)
)
plt.style.use("dark_background")
plt.show()
```

```python
# plt.figure(figsize=(20, 20))
(pd.crosstab(df["STATE_NAME"], df["PMGSY_SCHEME"], normalize="index") * 100).plot(
    kind="bar", stacked=True, figsize=(14, 5)
)
plt.style.use("bmh")
plt.grid(True)
plt.show()
```

```python
(pd.crosstab(df["STATE_NAME"], df["PMGSY_SCHEME"], normalize="index") * 100).plot(
    kind="line", figsize=(18, 5)
)
plt.style.use("dark_background")
plt.show()
```
