```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv(
    r"https://github.com/datasciencedojo/datasets/blob/master/titanic.csv?raw=true"
)
```

```python
from sklearn.impute import KNNImputer
```

```python
kn = KNNImputer(n_neighbors=5, weights="distance", metric="nan_euclidean")
```

```python
df["Age_New"] = kn.fit_transform(df[["Age"]])
```

```python
df.isna().sum()
```

```python
df["Age_New"].plot(kind="kde", label="After")
df["Age"].plot(kind="kde", label="Before")
plt.legend()
plt.show()
```
