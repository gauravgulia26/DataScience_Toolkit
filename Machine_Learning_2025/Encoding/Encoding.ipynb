```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder, LabelEncoder
```

```python
df = sns.load_dataset("titanic")
```

```python
df.replace({"survived": {0: "No", 1: "Yes"}}, inplace=True)
```

```python
# Survied -> LabelEncoder
# Class -> OrdinalEncoder
# Everthing Else -> OneHotEncoder
df.head()
```

```python
trf = [
    ("Ordinal", OrdinalEncoder(), [8]),
    (
        "Onehot",
        OneHotEncoder(drop="first", sparse_output=False, dtype=int),
        [2, 7, 9],
    ),
]
```

```python
transformations = ColumnTransformer(transformers=trf, remainder="passthrough")
```

```python
data = transformations.fit_transform(df)
```

```python
clm = transformations.get_feature_names_out()
```

```python
ml_df = pd.DataFrame(data=data, columns=clm)
```

```python
ml_df["remainder__survived"] = lc.fit_transform(ml_df["remainder__survived"])
```
