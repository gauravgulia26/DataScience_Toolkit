```python
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=500, n_features=4, noise=1, random_state=42)
```

```python
df = pd.DataFrame(X, columns=["f1", "f2", "f3", "f4"])
```

```python
plt.scatter(df["f4"], y)
```

```python
df = pd.concat([df, pd.Series(y)], axis=1)
```

```python
import seaborn as sns

sns.heatmap(df.corr())
```
